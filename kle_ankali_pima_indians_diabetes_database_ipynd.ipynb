```python
df.columns=df.columns.str.strip()
```

```python
(df == 0).any()
```

```python
df.columns = df.columns.str.strip().str.lower()
```

```python
#columns where 0 is not a value medical value

invalid_zero_cols = [
    'glucose',
    'bloodpressure',
    'skinthickness',
    'insulin',
    'bmi'
    ]

 #replace 0 with nan
df[invalid_zero_cols] = df[invalid_zero_cols].replace(0, np.nan)
```

```python
df.isnull().sum()
```

```python
#impute mising value

for col in invalid_zero_cols:
  df[col].fillna(df[col].median(), inplace=True)

  #mediam represents: a typical partient not too high not too low not
#influenced by outliers
```

```python
#feature target sepration

#feature
x= df.drop(columns = ['outcome'])

#target
y = df['outcome']
```

```python
x.columns
```

```python
y.name
```

```python
#train and test split

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
    )

```

```python
x_train.shape,x_test.shape
```

```python
y.value_counts(normalize=True)
```

```python
y_train.value_counts(normalize=True)
```

```python
y_test.value_counts(normalize=True)
```

```python
#FEATURES SCALLING- preprocessing

from sklearn.preprocessing import StandardScaler

scaler= StandardScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
# ere using logestic regression classifier because it is a binary classification problem,strong bseline for medical data
from sklearn.linear_model import LogisticRegression
log_reg=LogisticRegression()
log_reg.fit(x_train,y_train)

# here how the model trained, uses scaled features, can implement safely
```

```python
#predictions

y_pred = log_reg.predict(x_test)
```

```python
y_pred[:10]
```

```python
log_reg.score(x_test,y_test)

# accuary of test
```

```python
# accuary of train

log_reg.score(x_train,y_train)
```

```python
# linear regression probability prediction
y_prob = log_reg.predict_proba(x_test)
```

```python
y_prob[:20]
```

```python
y_prob_diabetic = y_prob[:,1]
y_prob_diabetic[:10]
```

```python
# confusion matrix

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test,y_pred)
cm
# [TN FP]
# [FN TP]

# TN - true negtive - actually non diabteic
# FP - false positive - actually non diabteic but predicted diabetic
# FN - false negtive - actually diabetic but predicted non diabetic
# TP - true positive - actually diabetic
```

```python
# classification report

from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))

# here we will see about precision, recall, f1 score and support
# precision: how correct the model was when it said 0 or 1
# recall: how many real 0 and 1 it found
# f1:balance score of precision and recall
# support : how many actual 0 and 1 patients

```

```python
#example for weighted avg
# weighted avg is avg of metrics where each class is weighted by its number of simple
# classes with more simples influence the final score more
# than the smaller classes

# next model we using here is SVM Support Vector
# machine, it finds a strong, margin based decisions
# boundary that often improves recall on small,
# well scalled datasets

# the main reason to use is dataset is small in size,
# we did the scaling, chance are high to get better
# recall
```
